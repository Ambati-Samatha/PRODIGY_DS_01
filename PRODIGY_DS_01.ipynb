```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
# Load the dataset
df = pd.read_csv('Dataset1.csv')
```

```python
# Convert 'DOB' column to datetime
df['DOB'] = pd.to_datetime(df['DOB'], errors='coerce') 
```

```python
# Calculate ages from birthdates
df['Age'] = (pd.to_datetime('today') - df['DOB']).dt.days // 365.25
```

```python
# Group data by gender, employment status, and recruitment source
gender_employment_source_counts = df.groupby(['Sex', 'EmploymentStatus', 'RecruitmentSource']).size().unstack()
```

```python
# Create a bar chart
fig, ax = plt.subplots(figsize=(10, 6))
gender_employment_source_counts.plot(kind='bar', stacked=True, ax=ax)

ax.set_xlabel('Gender and Employment Status')
ax.set_ylabel('Count')
ax.set_title('Distribution of Employment Status, Recruitment Source by Gender')
ax.legend(bbox_to_anchor=(1.05, 1))

plt.tight_layout()
plt.show()
```
